```python
import pandas as pd
import json

# Step 1: Load your reference JSON file
with open('/content/state_district_reference.json', 'r', encoding='utf-8') as f:
    reference_data = json.load(f)


# Step 3: Create a simple validation function
def is_valid_district(state, district):
    """
    Check if district exists in the given state's district list

    Args:
        state (str): State name (must match JSON key exactly)
        district (str): District name to validate

    Returns:
        bool: True if district is in that state's list, False otherwise
    """
    # Get the districts list for this state from JSON
    if state in reference_data:
        valid_districts = reference_data[state]['districts']
        # Check if district is in the list
        return district in valid_districts
    return False

# Step 4: Apply validation to each row
df['is_valid'] = df.apply(
    lambda row: is_valid_district(row['state'], row['district']),
    axis=1
)

# Step 5: Get rows where matching didn't happen (invalid rows)
invalid_rows = df[df['is_valid'] == False]

# Step 6: Display results
print("=" * 70)
print("DISTRICT VALIDATION REPORT")
print("=" * 70)
print(f"\nTotal rows in dataset: {len(df)}")
print(f"Valid matches (district found in state's list): {df['is_valid'].sum()}")
print(f"INVALID/MISSPELLED rows (matching didn't happen): {len(invalid_rows)}")
print(f"Percentage of invalid rows: {(len(invalid_rows)/len(df)*100):.2f}%")
print("=" * 70)

# Step 7: Show sample of invalid records
if len(invalid_rows) > 0:
    print("\nSample of INVALID records (first 10):")
    print(invalid_rows[['state', 'district']].head(10).to_string(index=False))

    # Breakdown by state
    print("\n\nINVALID records breakdown by STATE:")
    print(invalid_rows['state'].value_counts().to_string())

# Step 8: Decide whether to drop based on count
print("\n" + "=" * 70)
if len(invalid_rows) >= 2000 and len(invalid_rows) <= 3000:
    print(f"✓ Invalid rows count: {len(invalid_rows)} (within 2k-3k range - SAFE TO DROP)")

    # Drop invalid rows and create cleaned dataset
    df_cleaned = df[df['is_valid'] == True].drop('is_valid', axis=1).reset_index(drop=True)

    print(f"✓ Original dataset shape: {df.shape}")
    print(f"✓ Cleaned dataset shape: {df_cleaned.shape}")
    print(f"✓ Rows dropped: {len(df) - len(df_cleaned)}")

    # Save cleaned data
    df_cleaned.to_csv('your_dataset_cleaned.csv', index=False)
    print(f"\n✓ Cleaned dataset saved to 'your_dataset_cleaned.csv'")

elif len(invalid_rows) < 2000:
    print(f"⚠ Invalid rows count: {len(invalid_rows)} (less than 2000)")
    print("  Suggestion: Consider fixing these records instead of dropping")

elif len(invalid_rows) > 3000:
    print(f"✗ Invalid rows count: {len(invalid_rows)} (more than 3000)")
    print("  Suggestion: Data quality issue - manual review needed before dropping")

# Step 9: Export invalid records for reference
if len(invalid_rows) > 0:
    invalid_rows[['state', 'district']].to_csv('invalid_records.csv', index=False)
    print(f"✓ Invalid records exported to 'invalid_records.csv' for review")

print("=" * 70)
```
